````python
station1_crit = [44]
station_44 = data.loc[data['Stn Id'].isin(station1_crit)]
station44 = station_44[['Date','Precip (in)','Avg Air Temp (F)','Avg Rel Hum (%)']]
station44
````

````python
station44_new = station44.copy()
````

````python
station44_new['date'] = pd.to_datetime(station44_new["Date"]).dt.strftime("%Y%d%m")
print( station44_new )
#converting the Date column into string format in order to group by months
````

````python
station44_new['date'] = station44_new['date'].astype(str)
station44_new['month'] = station44_new['date'].str.slice(start=6,stop=8)
station44_new.head()
````

````python
grouped = station44_new.groupby('month')
print(len(grouped))
````

````python
list(grouped.groups.keys())[:5]
````

````python
for key, group in grouped:
    print("Key:\n", key)
    print("\nFirst rows of data in this group:\n", group.head())

    break
````

````python
cols = 'Avg Air Temp (F)'
station44_final = station44_new.groupby('month')[cols].mean().reset_index()
col_hum = 'Avg Rel Hum (%)'
station44_final['Mean Hum']= 0
Mean_Hum = station44_new.groupby('month')[col_hum].mean().reset_index()
station44_final['Mean Hum']= Mean_Hum['Avg Rel Hum (%)']
col_precip = 'Precip (in)'
station44_final['Mean Precip'] = 0
Mean_precip = station44_new.groupby('month')[col_precip].mean().reset_index()
station44_final['Mean Precip']=Mean_precip['Precip (in)']
station44_final
````
